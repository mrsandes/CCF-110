```python
# 1

valor = float(input("Digite o valor gasto: "))

print(f"Gorjeta: {valor * 0.10}")
print(f"Valor final: {(valor * 1.10):.2f}")
```

```python
# 2

numero = int(input("Digite o numero: "))

print(f"Sucessor: {numero + 1}")
print(f"Antecessor: {numero - 1}")
```

```python
# 3

chuva = float(input("Digite a quantidade de chuva em pol: "))

print(f"Chuva em mm: {chuva * 25.4:.2f}")
```

```python
# 4

base = float(input("Digite a base do retangulo: "))
altura = float(input("Digite a altura do retangulo: "))

print(f"Perímetro: {2 * (base + altura):.2f}")
print(f"Área: {base * altura:.2f}")
```

```python
# 5

a = int(input("primeiro valor: "))
b = int(input("segundo valor: "))
c = int(input("terceiro valor: "))

print(f"soma dos quadrados: {a**2 + b**2 + c**2}")
```

```python
# 6

R = float(input("Digite o raio: "))
H = float(input("Digite a altura: "))

V = 3.14159 * R**2 * H

print(f"Volume: {V:.2f}")
```

```python
# 7

a1 = float(input("Digite o primeiro valor: "))
r = float(input("Digite a razão: "))
n = int(input("Digite o n referente ao termo desejado: "))

print(f"o termo {n} é: {a1 + (n - 1) * r:.2f}")
```

```python
# 8

a1 = float(input("Digite o primeiro valor: "))
q = float(input("Digite a razão: "))
n = int(input("Digite o n referente ao termo desejado: "))

print(f"o termo {n} é: {a1 * q**(n - 1):.2f}")
```

```python
# 9

tempo = float(input("Digite o tempo gasto em horas: "))
velocidade = float(input("Digite a velocidade média em km/h: "))
distancia = tempo * velocidade

print(f"Distância percorrida: {distancia:.2f}")
print(f"litros de gasolina gastos: {distancia / 12:.2f}")
```

```python
# 10

custo = float(input("Digite o custo: "))
consumidor = custo + custo * 0.28 + custo * 0.45

print(f"custo para o consumidor: {consumidor:.2f}")
```
