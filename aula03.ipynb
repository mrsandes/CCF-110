```python
# 1045

a, b, c = list(map(float, input("").split()))

if a < b:
  x = a
  a = b
  b = x
    
if a < c:
  x = a
  a = c
  c = x

if a >= b + c:
  print("NAO FORMA TRIANGULO")

else:    
  if a**2 == b**2 + c**2:
    print("TRIANGULO RETANGULO")
      
  elif a**2 > b**2 + c**2:
    print("TRIANGULO OBTUSANGULO")
      
  elif a**2 < b**2 + c**2:
    print("TRIANGULO ACUTANGULO")
      
  if a == b == c:
    print("TRIANGULO EQUILATERO")
      
  if (a == b and a != c) or (a == c and a != b) or (b == c and b != a):
    print("TRIANGULO ISOSCELES")    
```

```python
# 1046

start, end = list(input("").split())
start = int(start)
end = int(end)

if start == end:
  print("O JOGO DUROU 24 HORA(S)")
  
else:
  if end > start:
    time = end - start
  
  else:
    time = 24 - start + end
  
  print(f"O JOGO DUROU {time} HORA(S)")
```

```python
# 1048

salary = float(input(""))

if salary < 400.01:
  rate = 15
    
elif salary < 800.01:
  rate = 12
    
elif salary < 1200.01:
  rate = 10
    
elif salary < 2000.01:
  rate = 7

else:
  rate = 4
    
print(f"Novo salario: {salary * (1 + rate / 100):.2f}")
print(f"Reajuste ganho: {salary * (1 + rate / 100) - salary:.2f}")
print(f"Em percentual: {rate} %")
```
