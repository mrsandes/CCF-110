```python
# 1019

seconds = int(input(""))

hours = seconds // 3600
minutes = (seconds - (hours * 3600)) // 60
seconds -= ((hours * 3600) + (minutes * 60))

print(f"{hours}:{minutes}:{seconds}")
```

```python
# 1071

x = int(input(""))
y = int(input(""))
sum = 0

for i in range(x - 1, y, -1):
  if i % 2 != 0:
    sum += i
    
print(sum)
```

```python
# 1101

M, N = map(int, input().split())

while M > 0 and N > 0:
  sum = 0
  
  if M > N:
    M, N = N, M
      
  for i in range(M, N + 1):
    print(f"{i} ", end="")
    sum += i

  print(f"Sum={sum}")
  
  M, N = map(int, input().split())
```

```python
# 1142

N = int(input(""))
i = 1

while N > 0:
  if i % 4 == 0:
    print("PUM")
    N -= 1
        
  else:
    print(f"{i} ", end="")
    
  i += 1
```
